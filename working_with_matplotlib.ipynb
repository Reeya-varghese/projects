```python
import matplotlib.pyplot as plt
import numpy as np

x1=np.array([6,2,7,11])
y1=np.array([3,8,1,10])
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}
plt.plot(y1,x1)
plt.title("patient data",fontdict=font1)
plt.xlabel("avg pulse",fontdict=font2)
plt.ylabel("calorie burnage",fontdict=font2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1=np.array([6,2,7,11])
y1=np.array([3,8,1,10])
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}
plt.plot(y1,x1)
plt.title("patient data",fontdict=font1,loc='left')
plt.xlabel("avg pulse",fontdict=font2)
plt.ylabel("calorie burnage",fontdict=font2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1=np.array([6,2,7,11])
y1=np.array([3,8,1,10])
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}
plt.plot(y1,x1)
plt.title("patient data",fontdict=font1)
plt.xlabel("avg pulse",fontdict=font2)
plt.ylabel("calorie burnage",fontdict=font2)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1=np.array([6,2,7,11])
y1=np.array([3,8,1,10])
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}
plt.plot(y1,x1)
plt.title("patient data",fontdict=font1)
plt.xlabel("avg pulse",fontdict=font2)
plt.ylabel("calorie burnage",fontdict=font2)
plt.grid(color='green',ls='--',lw=0.5)
plt.show()
```

```python
titles = []
texts = []

def extract_and_save_text(url, title, text, url_id):
    file_name = f"{url_id}.txt"

    with open(file_name, "w", encoding="utf-8") as file:
        file.write(text)

    titles.append(title)
    texts.append(text)
```

```python
def extract_article_text(url, url_id):
    try:

        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        header = soup.find(class_='td-header-template-wrap')
        if header:
            header.decompose()

        footer = soup.find(class_='td-footer-template-wrap')
        if footer:
            footer.decompose()

        title = soup.title.text.strip() if soup.title else "Untitled"


        article_text = ""
        for paragraph in soup.find_all('p'):
            article_text += paragraph.text + "\n"

        return url,url_id,title, article_text

    except requests.exceptions.HTTPError as e:
        if e.response.status_code == 404:
            print(f"404 Error: {url}")
            return url, url_id, "Page Not Found", "None"
        else:
            print(f"Error processing URL {url}: {e}")
            return url, url_id, None, None

    except Exception as e:
        print(f"Error processing URL {url}: {e}")
        return url, url_id, None, None,
```

```python
def load_master_dictionary(folder_path='/content/drive/MyDrive/MasterDictionary'):
    positive_words = set()
    negative_words = set()


    positive_file_path = os.path.join(folder_path, 'positive-words.txt')
    with open(positive_file_path, 'r', encoding='latin-1') as file:
        positive_words.update(file.read().split())


    negative_file_path = os.path.join(folder_path, 'negative-words.txt')
    with open(negative_file_path, 'r', encoding='latin-1') as file:
        negative_words.update(file.read().split())

    return positive_words, negative_words
```

```python
positive_words, negative_words = load_master_dictionary()
```

```python
def calculate_sentiment_scores(tokens):

    positive_score = sum(1 for token in tokens if token in positive_words)


    negative_score = sum(-1 for token in tokens if token in negative_words)


    negative_score = -1 * negative_score

    return positive_score, negative_score
```

```python
def break_sentences(text):
    nlp = spacy.load('en_core_web_sm')
    doc = nlp(text)
    return list(doc.sents)

def word_count(text):
    words = nltk.word_tokenize(text)
    words = [word for word in words if word.isalnum()]
    stop_words = set(stopwords.words('english'))
    words = [word.lower() for word in words if word.lower() not in stop_words]
    word_count = len(words)

    return word_count

def sentence_count(text):
    sentences = break_sentences(text)
    return len(sentences)

def avg_sentence_length(text):
    words = word_count(text)
    sentences = sentence_count(text)
    average_sentence_length = float(words / sentences)
    return average_sentence_length
```

```python

def syllables_count(word):
    exceptions = ["es", "ed"]
    for exception in exceptions:
        if word.endswith(exception):
            return max(1, count_vowels(word) - 1)

    return max(1, count_vowels(word))
```

```python

def count_vowels(word):
    vowels = "aeiouy"
    count = 0
    is_vowel = False

    for char in word:
        char_lower = char.lower()
        if char_lower in vowels:
            if not is_vowel:
                count += 1
                is_vowel = True
        else:
            is_vowel = False

    return count
```

```python
def avg_syllables_per_word(text):
    words = text.split()
    syllable_count_list = [syllables_count(word) for word in words]
    return (syllable_count_list)
```

```python
sample_text = "This is a sample text. It contains some words with multiple syllables like 'exception' and 'handled'."
syllable_count=avg_syllables_per_word(sample_text)
print("Syllable Count per Word:", sum(syllable_count)/len(syllable_count))
```

```python
def complex_words(text):

    nlp = spacy.load('en_core_web_sm')
    doc = nlp(text)

    words = []
    sentences = break_sentences(text)
    for sentence in sentences:
        words += [str(token) for token in sentence]

    complex_words_set = set()

    for word in words:
        syllable_count = syllables_count(word)
        if word not in nlp.Defaults.stop_words and syllable_count >= 2:
            complex_words_set.add(word)

    return len(complex_words_set)
```

```python
def gunning_fog(text):
    per_complex_words = (complex_words(text) / word_count(text) * 100) + 5
    grade = 0.4 * (avg_sentence_length(text) + per_complex_words)
    return grade
```

```python
import re

def count_personal_pronouns(text):
    personal_pronouns = ["I", "we", "my", "ours", "us"]
    pattern = re.compile(r'\b(?:' + '|'.join(personal_pronouns) + r')\b', flags= re.IGNORECASE)
    matches = re.findall(pattern, text)
    matches1 = [match for match in matches if match.lower() != 'us']
    total = len(matches1)
    return total
```

```python
sample_text = "This i a sample my text. It contains my we ours us some words with multiple syllables like 'exception' and 'handled'."
personal_count=count_personal_pronouns(sample_text)
print("personal pronoun Count per Word:", personal_count)
```

```python
def average_word_length(text):
    words = text.split()
    total_characters = sum(len(word) for word in words)
    total_words = len(words)
    if total_words > 0:
        avg_word_length = total_characters / total_words
    else:
        avg_word_length = 0

    return avg_word_length
```

```python
for index, row in extract_data.iterrows():
    title = row['Title']
    text = row['Text']

    avg_sent_len = avg_sentence_length(text)
    per_complex_words1 = (complex_words(text) / word_count(text) * 100) + 5
    fog_index = gunning_fog(text)
    Avg_words_per_sentence = word_count(text) / sentence_count(text)
    complex_word_count = complex_words(text)
    word_count1 = word_count(text)
    syllable_per_word = sum(avg_syllables_per_word(text))/len(avg_syllables_per_word(text))
    personal_pronouns = count_personal_pronouns(text)
    avg_word_length = average_word_length(text)

    Final_output.at[index, 'AVG_Sentence_Length'] = avg_sent_len

    Final_output.at[index, 'Percentage of complex words'] = per_complex_words1

    Final_output.at[index, 'Fog_index'] = fog_index

    Final_output.at[index, 'Average number of words per sentence'] = Avg_words_per_sentence

    Final_output.at[index, 'complex word count'] = complex_word_count

    Final_output.at[index, 'word count'] = word_count1

    Final_output.at[index, 'syllable per word'] = syllable_per_word

    Final_output.at[index, 'personal_pronouns'] = personal_pronouns

    Final_output.at[index, 'average word length'] = avg_word_length

```
